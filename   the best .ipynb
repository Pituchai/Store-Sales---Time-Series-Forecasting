```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error  # Don't forget this import!

# Assuming 'sales' is the target variable
y = merged_df['sales'] 
sales_features = ['id', 'store_nbr', 'family', 'onpromotion', 'dcoilwtico','city','type','cluster','transactions']
X = merged_df[sales_features]

from sklearn.model_selection import train_test_split

# Split the data into training and validation sets
X_train, X_val, Y_train, Y_val = train_test_split(X, y, test_size=0.3, random_state=0)

# Train a DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
train_model = DecisionTreeRegressor(random_state=0)
train_model.fit(X_train, Y_train)

# Make predictions on the validation set
predictions = train_model.predict(X_val)

# Evaluate the model
mse = mean_squared_error(Y_val, predictions)
print(f"Mean Squared Error on Validation Set: {mse}")

# Now, if you want to make predictions on a new dataset (e.g., merged_test)
# Assuming 'merged_test' has the same features as 'X'
predictions_test = train_model.predict(merged_test[sales_features])

# Create a DataFrame for submission
result_df_no_grid = pd.DataFrame({'id': merged_test['id'], 'sales': predictions_test})
result_df_no_grid.head()  # Check the first few rows of the result


```

## Decision Tree with Grid

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# Assuming 'sales' is the target variable
y = merged_df['sales'] 
sales_features = ['id', 'store_nbr', 'family', 'onpromotion', 'dcoilwtico', 'city']
X = merged_df[sales_features]

# Split the data into training and validation sets
X_train, X_val, Y_train, Y_val = train_test_split(X, y, test_size=0.3, random_state=0)

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(random_state=0)

# Initialize GridSearchCV
grid_search = GridSearchCV(tree_reg, param_grid, cv=5, scoring='neg_mean_squared_error')

print("Training the grid search model...")

# Fit the grid search to the data
grid_search.fit(X_train, Y_train)

print("Grid search completed.")

# Get the best hyperparameters
best_params = grid_search.best_params_
print(f"Best hyperparameters: {best_params}")

# Train a DecisionTreeRegressor with the best hyperparameters
best_tree_reg = DecisionTreeRegressor(random_state=0, **best_params)

print("Training the model with the best hyperparameters...")

best_tree_reg.fit(X_train, Y_train)

print("Model training completed.")

# Make predictions on the validation set
predictions = best_tree_reg.predict(X_val)

# Evaluate the model
mse = mean_squared_error(Y_val, predictions)
print(f"Mean Squared Error on Validation Set: {mse}")

print("Making predictions on the test set...")

# Now, if you want to make predictions on a new dataset (e.g., merged_test)
# Assuming 'merged_test' has the same features as 'X'
predictions_test = best_tree_reg.predict(merged_test[sales_features])

print("Prediction on the test set completed.")

# Create a DataFrame for submission
result_df = pd.DataFrame({'id': merged_test['id'], 'sales': predictions_test})
result_df.head()  # Check the first few rows of the result
```
